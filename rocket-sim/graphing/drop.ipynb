```python
import math
import numpy as np
import matplotlib.pyplot as plt
```


Kode hentet fra: https://stackoverflow.com/questions/71542662/python-plotting-free-fall-with-drag

Formel brukt (selv utviklet):
#### $\frac{-g\cdot m + \frac{ rho \cdot |v^2|}{2} \cdot Cd \cdot A}{mass}$

```python

mass = 0.208  # kg
g = 9.81    # m/s/s
c_1 = 0.8
rho_0 = 1.229
h_n = 10400
area = 0.01045
dt = 0.01     # reduce delta t to improve computation speed (result not impacted)
t = np.arange(0, 20, dt)  # extend time to 600 seconds

x = np.zeros(len(t))
v = np.zeros(len(t))
a = np.zeros(len(t))
rho = np.zeros(len(t))

x[0] = 100
v[0] = 0
a[0] = -g

for i in range(1, len(t)):
    rho[i] = rho_0 * np.exp(-x[i-1]/h_n)
    a[i] = -g + c_1*rho[i]*area*(v[i-1]**2)/mass
    v[i] = v[i-1] + a[i-1]*dt
    x[i] = x[i-1] + v[i-1]*dt
```

# Plots
hello

```python
plt.plot(t, x)
plt.xlabel('Time (s)')
plt.ylabel('Height (m)')
```

```python
plt.plot(t, v)
plt.xlabel('Time (s)')
plt.ylabel('Velocity (m/s)')
```

```python
plt.plot(t, a)
plt.xlabel('Time (s)')
plt.ylabel('Acceleration (m/s/s)')
```
